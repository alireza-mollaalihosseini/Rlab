```r
vac <- 17411
pla <- 17511
vac.pos <- 8
pla.pos <- 162
n.total <- vac + pla
#theta <- vac.pos/(vac.pos+pla.pos)
adj.vac <- vac.pos * ((n.total/2)/vac)
adj.pla <- pla.pos * ((n.total/2)/pla)
p <- seq(0,1,len=2001)
#post <- dbeta(x=p, shape1= alpha + adj.vac, shape2= beta + adj.pla)
```

```r
p.efficacy <- 0.3
theta <- (1-p.efficacy)/(2-p.efficacy)
beta <- 1
alpha <- -(1-1/(1-theta))

post <- dbeta(x=p, shape1= alpha + adj.vac, shape2= beta + adj.pla)
#f <- function(x){ 100*( 1 - (x /(1 - x))) }
#post <- 1 - (post.theta/(1-post.theta))
post.1 <- as.double(post)
post.2 <- as.double(as.double(100-200*post.1)/as.double(1-post.1))
```
